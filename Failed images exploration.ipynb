```python
import pandas as pd

fieldnames = ['imgname', 'center', 'scale', 'res', 'rot', 'ul', 'br', 'pad', 'old_x', 'old_y', 'new_x', 'new_y']

df = pd.read_csv('notebooks/errors_v1.csv', header=None, names=fieldnames)
```

```python
columns_to_convert = ['old_x', 'old_y', 'new_x', 'new_y']

new_df = pd.DataFrame()

for column_name in columns_to_convert:
    print(column_name)
    column_tuple = [literal_eval(row) for row in df[column_name]]
    lb_column = [x[0] for x in column_tuple]
    ub_column = [x[1] for x in column_tuple]
    new_df[str(column_name) + '_lb'] = lb_column
    new_df[str(column_name) + '_ub'] = ub_column
```

```python
def break_tuples(column):
    column_tuple = [literal_eval(row) for row in column]
    lb_column = [x[0] for x in column_tuple]
    ub_column = [x[1] for x in column_tuple]
    return lb_column, ub_column
```

```python
columns_to_convert = ['old_x', 'old_y', 'new_x', 'new_y']

new_df = pd.DataFrame()

for name in columns_to_convert:
    new_df[str(name) + '_lb'], new_df[str(name) + '_ub'] = break_tuples(df[name])
```
