# Investigating the outputs of imutils.crop

Code has been added to imutils.crop to make it output the data below to two .csv's: working.csv for examples that work and errors.csv for ValueError exceptions.

```python
import pandas as pd
from ast import literal_eval

fieldnames = ['imgname', 'center', 'scale', 'res', 'rot', 'ul', 'br', 'pad', 'old_x', 'old_y', 'new_x', 'new_y']

errors_csv = pd.read_csv('notebooks/errors.csv', header=None, names=fieldnames)
working_csv = pd.read_csv('notebooks/working.csv', header = None, names=fieldnames)
```

```python
def break_column(column):
    try:
        column_tuple = [literal_eval(row) for row in column]
        lb_column = [x[0] for x in column_tuple]
        ub_column = [x[1] for x in column_tuple]
        return lb_column, ub_column
    except SyntaxError: # Assuming it is a non-literal list
        column_list = [row.strip('\[\]').split(' ') for row in column]
        filtered_list = []
        for row in column_list: # A work-around for filtering out empty strings left from the above command
            filtered_row = []
            for e in row:
                if e != '':
                    filtered_row.append(e)
            filtered_list.append(filtered_row)        
        lb_column = [float(x[0]) for x in filtered_list]
        ub_column = [float(x[-1]) for x in filtered_list] # Since index [1] is an empty space
        return lb_column, ub_column
```

```python
columns_to_convert = ['old_x', 'old_y', 'new_x', 'new_y', 'center', 'ul', 'br']
```

```python
df_errors = pd.DataFrame()

for name in columns_to_convert:
    df_errors[str(name) + '_lb'], df_errors[str(name) + '_ub'] = break_column(errors_csv[name])
```

```python
df_working = pd.DataFrame()

for name in columns_to_convert:
    df_working[str(name) + '_lb'], df_working[str(name) + '_ub'] = break_column(working_csv[name])
```
